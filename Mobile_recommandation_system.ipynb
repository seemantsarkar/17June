```python
# convert data in lower case

def lower_case(text):
    return text.lower()

columns=df.columns
# print(columns)
for i in columns:
    df[str(i)]=df[str(i)].apply(lower_case)

df.head()
```

```python
# extract company name from details column

df["company"]=df["details"].str.split(" ",expand=True)[0]
```

```python
# extract model name from details column

df["model"]=df["details"].str.split(" ",expand=True)[1] +" "+ df["details"].str.split(" ",expand=True)[2]
```

```python
df.head()
```

```python
# extract color  from details column

df["color"]=df["details"].str.split(" ",expand=True)[3]
```

```python
df.head()
```

```python
import re     #ISME HAI ABHI

# create function for extract only digit from text
def integer(text):
    a=re.findall("\d",text)
    b="".join(a)
    return int(b)

#create function for remove Punctuation
def Punctuation_remove(text):
    if text !=" ":
        a=re.findall("\w",text)
        b="".join(a)
    return b
```

```python
df["price"]=df["price"].astype(str)
df["color"]=df["color"].astype(str)
```

```python
# apply function

df["price"]=df["price"].apply(integer)
df["color"]=df["color"].apply(Punctuation_remove)
```

```python
df.head()
```

```python
color=["sunrise","midnight","purple","starlight","blue","black","red"]
```

```python
def remove_color(text):
    if text in color:
        return text
    else :
        return "White"
```

```python
df["color"]=df["color"].apply(remove_color)
```

# import cosine_similarity from sklearn

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
def recommend(text):
    global df
    new_length=len(df)
    cols=list(df.columns)

    for i in cols:
        for k in range(new_length):
            if df[str(i)][k]==text:
                new=i
                break
    index = df[df[new] == text].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    index=[]
    for i in distances[1:6]:
        index.append(i[0])
    return df.iloc[index,:-1]
```

```python
recommend("vivo")
```

```python
reader=Reader(rating_scale=(1,5))
data=Dataset.load_from_df(df[['company','model','review']], reader)
```

```python
svd=SVD()
training=data.build_full_trainset()
svd.fit(training)
```

```python
def collabrative_filtering( company , top_n=5):
    testset = training.build_anti_testset()
    testset = filter(lambda x: x[0] == company , testset)
    predictions = svd.test(testset)
    predictions.sort(key=lambda x: x.est, reverse=True)
    recommendations = [prediction.iid for prediction in predictions[:top_n]]
    # print(recommendations)
    new_index=[]
    for i in ['galaxys23' ,'galaxya53' ,'zeroultra' ,'105dsplus' ,'muzik110']:
      # print(i)
      for j in range(len(df)):
        if df["model"][j]==i:
          new_index.append(j)
          break
    return new_index
    # return df.iloc[new_index , :]

```

```python
collabrative_filtering("apple")
```

```python
def hybrid_recommandation(model , top_n=5):
  content_value=content_based( model , top_n)
  collabrative_based=collabrative_filtering( model , top_n)
  total_index=list(set(content_value + collabrative_based ))
  return df.iloc[total_index , :-1 ]
```
